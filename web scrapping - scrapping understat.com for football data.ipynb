### Scrapping player data from understat.com using beautifulsoup

```python
# import packages and models

import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
# select url and store in a varaible
base_url = 'https://understat.com/player/'
player = str(input('Please enter player id: '))
url = base_url + player
```

```python
# View url
url
```

```python
# player data and some other data are stored inside script HTML tags.
# select all the scripts
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')
```

```python
# View scripts
scripts
```

```python
# convert to strings
strings = scripts[1].string
```

```python
strings
```

```python
# strip unnecessary symbols and get only JSON data 
ind_start = strings.index("('")+2 
ind_end = strings.index("')") 
json_data = strings[ind_start:ind_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')

#convert string to json format
data = json.loads(json_data)
```

```python
data
```

```python
# Select portion of data needed and store in lists
games =[]
goals =[]
shots = []
xG = []
assists =[]
xA = []
key_passes = []
team = []
yellow = []
red = []
npg = []
npgxG = []
xGChain = []
xGBuildup = []
data_season = data['season']


for index in range(len(data_season)):
    for key in data_season[index]:
        if key == 'games':
            games.append(data_season[index][key])
        if key == 'goals':
            goals.append(data_season[index][key])
        if key == 'shots':
            shots.append(data_season[index][key])
        if key == 'xG':
            xG.append(data_season[index][key])
        if key == 'assists':
            assists.append(data_season[index][key])
        if key == 'xA':
            xA.append(data_season[index][key])
        if key == 'key_passes':
            key_passes.append(data_season[index][key])
        if key == 'team':
            team.append(data_season[index][key])
        if key == 'yellow':
            yellow.append(data_season[index][key])
        if key == 'red':
            red.append(data_season[index][key])
        if key == 'npg':
            npg.append(data_season[index][key])
        if key == 'npgxG':
            npgxG.append(data_season[index][key])
        if key == 'xGChain':
            xGChain.append(data_season[index][key])
        if key == 'xGBuildup':
            xGBuildup.append(data_season[index][key])
        
```

```python
# convert data collected and store in diff. list to a data frame
col_names = ['games', 'goals','shots','xG','assists', 'xA', 'key_passes', 'team', 'yellow','red',
            'npg', 'npgxG', 'xGChain', 'xGBuildup']
df = pd.DataFrame([games,goals,shots,xG,assists,xA,key_passes,team,yellow,red,npg,npgxG,
                   xGChain,xGBuildup],index=col_names)

# Transpose the dataframe
df = df.T
```

### We can all the data are objects not int or float. We have to convert to int and float

```python
# Converting Multiple columns to int
df = df.astype({'games':'int','goals':'int', 'shots':'int', 'assists':'int', 
                'key_passes':'int', 'yellow':'int', 'red':'int'})
```

```python
# Convert multiple columns from sting to float
df = df.astype({'xG':'float','xA':'float','npg':'float', 'npgxG': 'float', 'xGChain':'float', 
                'xGBuildup':'float'})
```
